```

```

````python
import pandas as pd
from twd2latlon import TMToLatLon

fin = open('108.sel')

a1=[]
a2=[]
a3=[]
a4=[]
count = 0

c1 = TMToLatLon()

#mark="E_97={},N_97={},Lon={},Lat={}"

for line in fin:
    count += 1
    if count <= 2:
        continue
    e_97 = float(line[17:27])
    n_97 = float(line[28:39])
  
    lat, lon = c1.convert(e_97, n_97)
    

    
    a1+=[e_97]
    a2+=[n_97]
    a3+=[lat]
    a4+=[lon]
       
    mark = {"E_97":a1,"N_97":a2,"Lon":a3,"Lat":a4}
    print(mark)
    
df = pd.DataFrame(mark)
    
out_file = "test110207417.csv"
df.to_csv(out_file, index=False)

````
