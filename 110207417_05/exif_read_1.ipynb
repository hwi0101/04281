```python
import exifread
# Open image file for reading (binary mode)

tags_list = ['Image Orientation', 'GPS GPSLatitudeRef', 'GPS GPSLatitude', 'GPS GPSLongitudeRef', 
            'GPS GPSLongitude', 'GPS GPSAltitudeRef', 'GPS GPSAltitude',
            'EXIF DateTimeOriginal', 'EXIF ExifImageWidth', 'EXIF ExifImageLength']
path_name = 'images/IMG_20200912_172322.jpg'
f = open(path_name, 'rb')

# Return Exif tags
tags = exifread.process_file(f)
for k in tags:
    if k in tags_list:
        print('{}:{}'.format(k, tags[k]), type(tags[k]))
```
